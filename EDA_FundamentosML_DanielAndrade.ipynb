Importacion de las librerias necesarias para el analisis de datos y visualizar graficamente dichos datos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Cargamos el archivo csv

```python
from google.colab import files
uploaded = files.upload()
for fn in uploaded.keys():
  name=fn
  df=pd.read_csv(name, sep = '\t')
```

Calculos de Tendencia Central:

Media:

```python
columnas_numericas = df.select_dtypes(include=np.number)
medias = columnas_numericas.mean()
tabla_medias = medias.to_frame().rename(columns={0: 'Media'})
print(tabla_medias)
```

Mediana:

```python
columnas_numericas = df.select_dtypes(include=np.number)
mediana = columnas_numericas.median()
tabla_mediana = mediana.to_frame().rename(columns={0: 'Mediana'})
print(tabla_mediana)
```

Histograma:

```python
plt.hist(df['Income'], bins=10)
plt.xlabel('Income')
plt.ylabel('Frecuencia')
plt.title('Histograma de Income')
plt.show()
```
